```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_FILE_BPDN_max = 'data/generated/BW_alphas-BPDN_10000_2024-04-07-12-43-32.npy'
DATA_FILE_BPDN_1 = 'data/generated/BW_alphas-BPDN-1iters_10000_2024-04-07-12-43-32.npy'
DATA_FILE_BPDN_3 = 'data/generated/BW_alphas-BPDN-3iters_10000_2024-04-07-12-43-32.npy'
```

```python
bpdn_max = np.load(DATA_FILE_BPDN_max)
bpdn_1 = np.load(DATA_FILE_BPDN_1)
bpdn_3= np.load(DATA_FILE_BPDN_3)
```

```python
def plot_comp(smpl):
    plt.plot(bpdn_max[smpl, :], label='max')
    plt.plot(bpdn_1[smpl, :], label='1')
    plt.plot(bpdn_3[smpl, :], label='3')
    plt.legend()
    plt.show()
```

```python
plot_comp(8000)
```

```python
plot_comp(8500)
```

```python
plot_comp(8950)
```
